```python
def calcular_media(nums):
    return sum(nums) / len(nums)

def calcular_media_geometrica(nums):
    product = 1
    for num in nums:
        product *= num
    return product ** (1/len(nums))

def solicitar_pesos(nums):
    pesos = []
    for num in nums:
        while True:
            try:
                peso = float(input(f'Digite o peso para o número {num}: '))
                pesos.append(peso)
                break
            except:
                print('Digite um peso válido.')
    return pesos

def calcular_media_ponderada(nums):
    pesos = solicitar_pesos(nums)
    return sum(nums[i] * pesos[i] for i in range(len(nums))) / sum(pesos)

def calcular_media_harmonica(nums):
    if 0 in nums:
        return None
    return len(nums) / sum(1/num for num in nums)

def variancia_amostral(nums):
    n = len(nums)
    media = calcular_media(nums)
    return sum((x - media)**2 for x in nums) / (n - 1)

def variancia_populacional(nums):
    n = len(nums)
    media = calcular_media(nums)
    return sum((x - media)**2 for x in nums) / n

def desvio_padrao_amostral(nums):
    return variancia_amostral(nums) ** 0.5

def desvio_padrao_populacional(nums):
    return variancia_populacional(nums) ** 0.5

def incerteza_media(nums):
    return desvio_padrao_amostral(nums) / (len(nums) ** 0.5)

def calcular_mediana(nums):
    sorted_nums = sorted(nums)
    n = len(sorted_nums)
    if n % 2 == 1:
        return sorted_nums[n // 2]
    else:
        middle1 = sorted_nums[(n - 1) // 2]
        middle2 = sorted_nums[n // 2]
        return (middle1 + middle2) / 2

def medias():
    nums = []
    
    while True:
        print('''\n1- Média Aritmética
2- Média Geometrica
3- Média Ponderada
4- Média Harmônica
5- Mediana
6- Variância Amostral
7- Variância Populacional
8- Desvio Padrão Amostral
9- Desvio Padrão Populacional
10- Incerteza da Média
11- Sair''')
        
        num = input('Digite um número ou aperte Q para sair: ').lower()
        if num == 'q':
            break
        try:
            nums.append(float(num))
        except ValueError:
            print('Digite um número válido.')
        
        opc = int(input('\nDigite aqui: '))

        if opc == 1:
            print(f'Média Aritmética: {calcular_media(nums)}')
        elif opc == 2:
            print(f'Média Geometrica: {calcular_media_geometrica(nums)}')
        elif opc == 3:
            print(f'Média Ponderada: {calcular_media_ponderada(nums)}')
        elif opc == 4:
            harmonica = calcular_media_harmonica(nums)
            if harmonica:
                print(f'Média Harmônica: {harmonica}')
            else:
                print("A média harmônica não pode ser calculada com um valor zero na lista.")
        elif opc == 5:
            print(f'Mediana: {calcular_mediana(nums)}')
        elif opc == 6:
            print(f'Variância Amostral: {variancia_amostral(nums)}')
        elif opc == 7:
            print(f'Variância Populacional: {variancia_populacional(nums)}')
        elif opc == 8:
            print(f'Desvio Padrão Amostral: {desvio_padrao_amostral(nums)}')
        elif opc == 9:
            print(f'Desvio Padrão Populacional: {desvio_padrao_populacional(nums)}')
        elif opc == 10:
            print(f'Incerteza da Média: {incerteza_media(nums)}')
        elif opc == 11:
            print('Saindo...')
            break


medias()
```

```python
import numpy as np
from scipy.stats import gmean, hmean

def calcular_media(nums):
    return np.mean(nums)

def calcular_media_geometrica(nums):
    return gmean(nums)

def solicitar_pesos(nums):
    pesos = []
    for num in nums:
        while True:
            try:
                peso = float(input(f'Digite o peso para o número {num}: '))
                pesos.append(peso)
                break
            except ValueError:
                print('Digite um peso válido.')
    return pesos

def calcular_media_ponderada(nums):
    pesos = solicitar_pesos(nums)
    return np.average(nums, weights=pesos)

def calcular_media_harmonica(nums):
    if 0 in nums:
        return None
    return hmean(nums)

def variancia_amostral(nums):
    return np.var(nums, ddof=1)

def variancia_populacional(nums):
    return np.var(nums)

def desvio_padrao_amostral(nums):
    return np.std(nums, ddof=1)

def desvio_padrao_populacional(nums):
    return np.std(nums)

def incerteza_media(nums):
    return desvio_padrao_amostral(nums) / np.sqrt(len(nums))

def calcular_mediana(nums):
    return np.median(nums)

def medias():
    nums = []
    
    while True:
        print('''\n1- Média Aritmética
2- Média Geometrica
3- Média Ponderada
4- Média Harmônica
5- Mediana
6- Variância Amostral
7- Variância Populacional
8- Desvio Padrão Amostral
9- Desvio Padrão Populacional
10- Incerteza da Média
11- Sair''')
        
        num = input('Digite um número ou aperte Q para sair: ').lower()
        if num == 'q':
            break
        try:
            nums.append(float(num))
        except ValueError:
            print('Digite um número válido.')
        
        opc = int(input('\nDigite aqui: '))

        if opc == 1:
            print(f'Média Aritmética: {calcular_media(nums)}')
        
        elif opc == 2:
            print(f'Média Geometrica: {calcular_media_geometrica(nums)}')
        
        elif opc == 3:
            print(f'Média Ponderada: {calcular_media_ponderada(nums)}')
        
        elif opc == 4:
            harmonica = calcular_media_harmonica(nums)
            if harmonica:
                print(f'Média Harmônica: {harmonica}')
            else:
                print("A média harmônica não pode ser calculada com um valor zero na lista.")
        
        elif opc == 5:
            print(f'Mediana: {calcular_mediana(nums)}')
        
        elif opc == 6:
            print(f'Variância Amostral: {variancia_amostral(nums)}')
        
        elif opc == 7:
            print(f'Variância Populacional: {variancia_populacional(nums)}')
        
        elif opc == 8:
            print(f'Desvio Padrão Amostral: {desvio_padrao_amostral(nums)}')
        
        elif opc == 9:
            print(f'Desvio Padrão Populacional: {desvio_padrao_populacional(nums)}')
        
        elif opc == 10:
            print(f'Incerteza da Média: {incerteza_media(nums)}')
        
        elif opc == 11:
            print('Saindo...')
            break

medias()
```

```python
import random
import numpy as np
from scipy.stats import gmean, hmean
import timeit

# Funções sem bibliotecas

def calcular_media_no_lib(nums):
    return sum(nums) / len(nums)

def calcular_media_geometrica_no_lib(nums):
    product = 1
    for num in nums:
        product *= num
    return product ** (1/len(nums))

def calcular_media_ponderada_no_lib(nums):
    pesos = [random.uniform(0.5, 1.5) for _ in nums]
    return sum(nums[i] * pesos[i] for i in range(len(nums))) / sum(pesos)

def calcular_media_harmonica_no_lib(nums):
    if 0 in nums:
        return None
    return len(nums) / sum(1/num for num in nums)

def calcular_mediana_no_lib(nums):
    sorted_nums = sorted(nums)
    n = len(sorted_nums)
    if n % 2 == 1:
        return sorted_nums[n // 2]
    else:
        middle1 = sorted_nums[(n - 1) // 2]
        middle2 = sorted_nums[n // 2]
        return (middle1 + middle2) / 2

# Funções com bibliotecas

def calcular_media_lib(nums):
    return np.mean(nums)

def calcular_media_geometrica_lib(nums):
    return gmean(nums)

def calcular_media_ponderada_lib(nums):
    pesos = [random.uniform(0.5, 1.5) for _ in nums]
    return np.average(nums, weights=pesos)

def calcular_media_harmonica_lib(nums):
    if 0 in nums:
        return None
    return hmean(nums)

def calcular_mediana_lib(nums):
    return np.median(nums)

# Comparação de desempenho

nums = [random.randint(0, 99) for _ in range(100000)]

print("Tempo sem bibliotecas:")
print("Média:", timeit.timeit(lambda: calcular_media_no_lib(nums), number=100))
print("Média Geométrica:", timeit.timeit(lambda: calcular_media_geometrica_no_lib(nums), number=100))
print("Média Ponderada:", timeit.timeit(lambda: calcular_media_ponderada_no_lib(nums), number=100))
print("Média Harmônica:", timeit.timeit(lambda: calcular_media_harmonica_no_lib(nums), number=100))
print("Mediana:", timeit.timeit(lambda: calcular_mediana_no_lib(nums), number=100))

print("\nTempo com bibliotecas:")
print("Média:", timeit.timeit(lambda: calcular_media_lib(nums), number=100))
print("Média Geométrica:", timeit.timeit(lambda: calcular_media_geometrica_lib(nums), number=100))
print("Média Ponderada:", timeit.timeit(lambda: calcular_media_ponderada_lib(nums), number=100))
print("Média Harmônica:", timeit.timeit(lambda: calcular_media_harmonica_lib(nums), number=100))
print("Mediana:", timeit.timeit(lambda: calcular_mediana_lib(nums), number=100))
```
