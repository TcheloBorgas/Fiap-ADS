```python
import time

#━━━━━━━━━❮menu❯━━━━━━━━━
def exibir_menu():
    print('-=' * 20)
    print('1 - Cadastrar loja')
    print('2 - Cadastrar produto')
    print('3 - Mostrar lojas')
    print('4 - Mostrar produtos')
    print('0 - Sair')
    print('-=' * 20)

    return int(input('Digite a opção: '))

#━━━━━━━━━❮Cadastro Loja❯━━━━━━━━━

def cadastrar_loja(lojas):
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('      Cadastro de Loja')
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('')
    
    id_loja = int(input('ID da Loja: '))
    nome_loja = input('Nome da Loja: ')

    lojas.append({'id': id_loja, 'nome': nome_loja})
    
    print('Loja cadastrada com sucesso!')
    print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')


#━━━━━━━━━❮Cadastro Produto❯━━━━━━━━━

def cadastrar_produto(produtos, lojas):
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('     Cadastro de Produto')
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('')

    id_produto = int(input('ID do Produto: '))
    nome_produto = input('Nome do Produto: ')
    print('')
    descricao_produto = input('Descrição do Produto: ')
    valor_produto = float(input('Valor do Produto: '))
    loja_id = int(input('ID da Loja: '))

    # Verifica se a loja existe
    loja_existente = False
    for loja in lojas:
        if loja['id'] == loja_id:
            loja_existente = True
            break
    
    if not loja_existente:
        print('Loja não encontrada. O produto não será cadastrado.')
        print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
        return

    produtos.append({'id': id_produto, 'nome': nome_produto, 'descricao': descricao_produto, 'valor': valor_produto, 'loja_id': loja_id})

    print('')
    print('Produto cadastrado com sucesso!')
    print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')

#━━━━━━━━━❮Mostrar Lojas❯━━━━━━━━━
def mostrar_lojas(lojas):
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('         Lista de Lojas')
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')

    for loja in lojas:
        print(f'{loja["id"]}: {loja["nome"]}')
    
    print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')

#━━━━━━━━━❮Mostrar Produtos❯━━━━━━━━━
def mostrar_produtos(produtos, lojas):
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    print('       Lista de Produtos')
    print('━━━━━━━━━━━━━❮◆❯━━━━━━━━━━━━━')
    for produto in produtos:
        # Busca o nome da loja pelo id
        nome_loja = ''
        for loja in lojas:
            if loja['id'] == produto['loja_id']:
                nome_loja = loja['nome']
                break

        print(f'{produto["id"]}: {produto["nome"]} - R${produto["valor"]:.2f} - Loja: {nome_loja}')
    
    print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')

#━━━━━━━━━❮Programa Principal❯━━━━━━━━━
lojas = []
produtos = []

while True:
    time.sleep(5)
    opcao = exibir_menu()

    if opcao == 1:
        cadastrar_loja(lojas)

    elif opcao == 2:
        cadastrar_produto(produtos, lojas)

    elif opcao == 3:
        mostrar_lojas(lojas)

    elif opcao == 4:
        mostrar_produtos(produtos, lojas)

    elif opcao == 0:
        print('Encerrando o programa...')
        break

    else:
        print('Opção inválida. Tente novamente.')
```
